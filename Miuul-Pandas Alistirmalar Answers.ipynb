```python
for col in df.columns:
    b = df[col].isnull().sum()
    print(f"Null values of {col} is: {b}")
```

```python
df.drop(columns=["who"], inplace=True)
df.head()
```

```python
df["deck"] = df["deck"].fillna(df["deck"].mode()[0])
df.head()
```

```python
df["age"] = df["age"].fillna(df["age"].median())
df.head()
```

```python
df.groupby(["pclass", "sex"]).agg({"survived": ["sum", "count", "mean"]})
```

```python
df["age_flag"] = df.loc[:, df.columns.str.contains("age")].apply(lambda x: 1 if x[0] < 30 else 0, axis=1)
df.head()
```

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
df = sns.load_dataset("tips")
df.head()
```

```python
df.groupby("time").agg({"total_bill": ["sum", "min", "max", "mean"]})
```

```python
df.groupby(["time", "day"]).agg({"total_bill": ["sum", "min", "max", "mean"]})
```

```python
df_new = df.loc[(df["time"] == "Lunch") & (df["sex"] == "Female"), ["total_bill", "tip", "day"]]
df_new.groupby("day").agg({"total_bill": ["sum", "min", "max", "mean"],
                           "tip": ["sum", "min", "max", "mean"]})
```

```python
df.loc[(df["size"] < 3) & (df["total_bill"] > 10), :].mean()
```

```python
total_bill_tip_sum = df["total_bill"] + df["tip"]
df["total_bill_tip_sum"] = total_bill_tip_sum
```

```python
df.head()
```

```python
df.sort_values(by=["total_bill_tip_sum"], ascending=False)
```
